# **Create Dummy DataFrame**

```python
# import library
import pandas as pd
```

```python
# Creating a dummy dataframe
data = {
    'Screen': ['Big', 'Big', 'Medium', 'Small', 'Small'],
    'Model': ['iPhone 12', 'Galaxy S20', 'Pixel 5', 'iPhone 11', 'Galaxy S10'],
    'Defect': ['Yes', 'No', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)
df
```

# **Create Dummy Variable with Pandas get_dummies()**

```python
# One-Hot Encoding using pandas
df_one_hot_pandas = pd.get_dummies(df, columns=['Model'], prefix='Model')
```

```python
df_one_hot_pandas
```

```python
# import function
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
```

```python
# One-Hot Encoding using scikit-learn
encoded_sklearn = ohe.fit_transform(df[['Model']])
df_one_hot_sklearn = pd.DataFrame(encoded_sklearn, columns=ohe.get_feature_names_out(['Model']))
```

```python
df_one_hot_sklearn
```
